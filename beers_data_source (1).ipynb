It seems the beer dataset contains data about the beer and the Breweries contains data about the breweries producing them.

the "abv" column is the alcoholic content by volume with 0 being no alcohol and 1 being pure alcohol, the "ibu" column is known as International bittering units, which describe how bitter a drink is. 
The "id" column is the unique id of each beer and the name column contains the name of the beer. The style has to do with the Beer style (lager, ale, IPA, etc.).
 the "brewery_id" column is the Unique identifier for the brewery that produces this beer; and can use to join with brewery info, finally, the "ounce" column is the Size of beer in ounces.

```python
%%sql

SELECT beers.ibu, beers.brewery_id, beers.beers_name, breweries.brewery_id, breweries.brewery_name
FROM beers
INNER JOIN breweries
ON beers.brewery_id = breweries.brewery_id
ORDER BY beers.ibu DESC
LIMIT 5
```

The beer with the beer name Bitter Bitch Imperial IPA and and produced by the Astoria Brewing Company (ibu = 138.0) is the most bitter alcoholic beer in this combined data sets.

```python
%%sql

SELECT be.ibu, be.brewery_id, be.beers_name, br.brewery_id, br.brewery_name
FROM beers AS be
INNER JOIN breweries AS br
ON be.brewery_id = br.brewery_id
ORDER BY be.ibu DESC
LIMIT 5
```

```python
%%sql

SELECT beers.brewery_id, breweries.state, breweries.brewery_name, beers.beers_name, COUNT(beers.beers_name) as num_of_beers_name
FROM beers
INNER JOIN breweries
ON beers.brewery_id = breweries.brewery_id
GROUP BY beers.brewery_id, breweries.state, breweries.brewery_name, beers.beers_name
ORDER BY num_of_beers_name DESC
LIMIT 3
```
